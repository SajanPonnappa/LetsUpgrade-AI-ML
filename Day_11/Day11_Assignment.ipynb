# Correlation between attrition and other variables(features)

```python
import pandas as pd
```

```python
df = pd.read_csv("general_data.csv")
```

```python
df.head()
```

### convert attirtion data to numerical values

```python
df['Attrition_numerical'] = pd.factorize(df.Attrition)[0]
```

```python
df.head()
```

```python
df.Attrition.value_counts()
```

```python
df.Attrition_numerical.value_counts()
```

```python
from scipy.stats import pearsonr
```

### Negative correlation between Attrition and Age

```python
stats,p = pearsonr(df.Attrition_numerical,df.MonthlyIncome)
print(stats,p)
```

### Negative correlation between Attrition and DistanceFromHome

```python
stats,p = pearsonr(df.Attrition_numerical,df.DistanceFromHome)
print(stats,p)
```

### Negative correlation between Attrition and JobLevel

```python
stats,p = pearsonr(df.Attrition_numerical,df.JobLevel)
print(stats,p)
```

### Negative correlation between Attrition and Monthly income

```python
stats,p = pearsonr(df.Attrition_numerical,df.MonthlyIncome)
print(stats,p)
```

### Positive correlation between Attrition and PercentSalaryHike

```python
stats,p = pearsonr(df.Attrition_numerical,df.PercentSalaryHike)
print(stats,p)
```

### Negative correlation between Attrition and StockOptionLevel

```python
stats,p = pearsonr(df.Attrition_numerical,df.StockOptionLevel)
print(stats,p)
```

### Negative correlation between Attrition and TrainingTimesLastYear
