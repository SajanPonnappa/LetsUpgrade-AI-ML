### Target variable is Attrition which is Binary Categorical

```python
data=pd.read_csv("general_data.csv")
data.head()
```

```python
data.info()
```

### Checking Null Columns

```python
data.isnull().sum()
```

### Filling Null Values

```python
data['TotalWorkingYears'] = data['TotalWorkingYears'].fillna(11.28) 
# 11.28 is the mean of TotalWorkingYears column 

data['NumCompaniesWorked'] = data['NumCompaniesWorked'].fillna(2)
# 2 is the median of NumCompaniesWorked
```

### Check if all values are filled

```python
data.isna().sum()
```

### Plotting histogram

```python
data.hist(figsize=(10,10))
plt.show()
```

### Inference from Histogram:
- Education, JobLevel, StockOptional Level are Categorical Variables which are in Encoded format
- Standard Hours, Employee Count, Employee ID are irrelevent columns

### Encoding text variarbles into number

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
data['Attrition'] = encoder.fit_transform(data['Attrition'])
data['BusinessTravel'] = encoder.fit_transform(data['BusinessTravel'])
data['Department'] = encoder.fit_transform(data['Department'])
data['Over18'] = encoder.fit_transform(data['Over18'])
data['MaritalStatus'] = encoder.fit_transform(data['MaritalStatus'])
data['JobRole'] = encoder.fit_transform(data['JobRole'])
data['Gender'] = encoder.fit_transform(data['Gender'])
data['EducationField'] = encoder.fit_transform(data['EducationField'])
```

### Drop irrelevant columns

```python
data = data.drop(['EmployeeCount','EmployeeID', 'Over18','StandardHours'], axis=1)
data.head()
```

### Splitting the dataset

```python
y = data.iloc[:,1:2]
y.head()
```

```python
x = data.drop(['Attrition'], axis=1)
x.head()
```

### Applying Logistic Regression

```python
import statsmodels.api as sm
```

```python
x1 = sm.add_constant(x)
```

```python
logistic = sm.Logit(y,x1)
```

```python
result = logistic.fit()
```

```python
result.summary()
```

2) Out of all these variables, namely Age, MaritalStatus, NumCompaniesWorked, TotalWorkingYears, TrainingTimesLastYear,YearsSinceLastPromotion and YearsWithCurrManager are most significant.
